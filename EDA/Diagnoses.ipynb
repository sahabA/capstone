    Let's just use these instead of using very detailed diagnoses

```python
def getDiagnoseClassification(row):
    
    #icd9code = row
    #print row
    row = str(row)
    if (row[0] is 'V' or row[0] is'E'):
        return "External Causes"
    #try:
    icd9code = int(row[0:3])
    if (icd9code <= 139):
        return "infectious and parasitic diseases"
    elif (icd9code <= 239):
        return "neoplasms"
    elif (icd9code <= 279):
        return "endocrine, nutritional and metabolic diseases, and immunity disorders"
    elif (icd9code <= 289):
        return "diseases of the blood and blood-forming organs"
    elif (icd9code <= 319):
        return "mental disorders"
    elif (icd9code <= 359):
        return "diseases of the nervous system"
    elif (icd9code <= 389):
        return "diseases of the sense organs"
    elif (icd9code <= 459):
        return "diseases of the circulatory systems"
    elif (icd9code <= 519):
        return "diseases of the respiratory system"
    elif (icd9code <= 579):
        return "diseases of the digestive system"
    elif (icd9code <= 629):
        return "diseases of the genitourinary system"
    elif (icd9code <= 679):
        return "complications of pregnancy, childbirth, and the puerperium"
    elif (icd9code <= 709):
        return "diseases of the skin and subcutaneous tissue"
    elif (icd9code <= 739):
        return "diseases of the musculoskeletal system and connective tissue"
    elif (icd9code <= 759):
        return "congenital anomalies"
    elif (icd9code <= 779):
        return "certain conditions originating in the perinatal period"
    elif (icd9code <= 799):
        return "symptoms, signs, and ill-defined conditions"
    elif (icd9code <= 999):
        return "injury and poisoning"

    
    #if icd9Code

diagnoses['diagnoses'] = diagnoses['ICD9_CODE'].map(getDiagnoseClassification)
```
